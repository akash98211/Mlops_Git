```python
to check the code editor 
command:: git config --global core.editor 

to set VIM as code editor 
command:: git config --global core.editor vim 
```

```python
revert 

Syntax: 
git revert commit_id 

when merge conflicts happens 
to abort 
command: git revert --abort

after resolving conflicts 
command: git revert --continue

to accept the incoming changes 
command : git revert --skip


```

## Rebase

```python
rebase is used to merge to combine the commits

Syntax: 
git rebase -i HEAD~3    

3= no of commits to merged into single commit
Steps: 
here vim editor will ask either to squash or fixup the previous commit 
```
